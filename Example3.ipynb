```python
import math
import  matplotlib.pyplot  as  plt

Vx=0
Vy=8.2

X1=0.47
Y1=0
t=0.001
r=1
X=10000

A=[]
B=[]
    
def planet(X,Vx,Vy,X1,Y1,t,r):
    for i in range (0,X):
        r=math.sqrt(X1*X1+Y1*Y1)
        Vx=Vx-((4*math.pi*math.pi*X1)/(r*r*r))*t
        Vy=Vy-((4*math.pi*math.pi*Y1)/(r*r*r))*t
        X1=X1+Vx*t
        Y1=Y1+Vy*t
        A.append(X1)
        B.append(Y1)
    return(A,B)
F=planet(10000,Vx,Vy,X1,Y1,t,r)
print(F)
```

```python
import  matplotlib.pyplot  as  plt 
plt.plot(F[0],  F[1],  "-" , [0] , [0] , 'ro') 
plt.axis([ -0.6 ,  0.6 ,  -0.6 ,  0.6 ]) 
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```
